```
import pandas as pd

data = pd.read_csv("/content/drive/My Drive/dataset/diabetes.csv")
len(data)
data
```

```
from sklearn.model_selection import train_test_split

x = data.iloc[:, 0:8]
y = data.iloc[:, 8]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)

x_train
```

```
from sklearn import svm
from sklearn.metrics import classification_report

clf = svm.SVC(gamma = 'scale')
clf = clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print (y_pred)

print(classification_report(y_pred, y_test))
```

```
from sklearn import tree

clf_dt = tree.DecisionTreeClassifier()
clf_dt = clf_dt.fit(x_train,y_train)
y_pred_dt = clf_dt.predict(x_test)
print (y_pred_dt)
print (y_test)
print(classification_report(y_pred_dt, y_test))
```

```
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3).fit(x_train)
y_pred_kmeans = kmeans.predict(x_test)
y_pred_kmeans
```

```
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(5, activation='relu', input_shape = (8,)))
model.add(Dense(5, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss = 'mean_squared_error', metrics=['accuracy'])
model.fit(x_train, y_train, validation_split=0.2,epochs=30)
```
